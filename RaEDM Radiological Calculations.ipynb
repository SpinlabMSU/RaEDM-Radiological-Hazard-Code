In the operation of the Radium EDM experiment's new 60 kV Heinzinger high voltage (HV) power supply (PSU), there could be a production of X-rays when there's a discharge between two components at different potentials.

Discharges in air are not expected to produce X-rays as electrons are apparently not accelerated to high enough energies to radiate Bremsstrahlung.

X-rays COULD be produced by a short burst of electrons striking one of the electrodes, while under vacuum. The resulting X-ray spectrum has a sharp and a broadband component. (**Sharp**: Spectral features with small wavelength range, **Broadband**: Spectral features with large wavelength range). 

### Kramer's Formula

The high energy part of the continuous Bremsstrahlung spectrum is approximated by Kramer's Formula:

## $\frac{d^2 N_{\gamma}}{dtdu}=\left[ \frac{I_0}{e} \right] \left[ \frac{2\alpha}{3\sqrt{3\ell}} \right] \left[ \frac{eV_0}{m_e c^2} \right] \left[ \frac{2Z(1-u)}{u} \right]$

Where the parameters involved have the following definitions:

* $N_\gamma$: The maximum number of X-rays
* $E$: The energy of the X-rays
* $u$: The relative energy of the X-rays, defined as $u = \frac{E}{E_0}$
* $E_0$: $E_0 = eV_0$
* $V_0$: The potential between the HV electrodes
* $e$: Elementary charge of an electron
* $I$: Electron beam current
* $\alpha$: Fine structure constant, $\alpha \approx \frac{1}{137}$
* $\ell$: Thickness of the "target", in this case I think its the thickness of the electrode
* $m_e$: electron mass
* $c$: The speed of light, 299,792,458 m/s
* $Z$: Atomic number of the target material

### Assumption 1: Isotropic Angular Distribution of X-Rays

We can assume that for X-Ray energies much less than 511 keV (annihilation radiation), the angular distribution of the X-rays is isotropic. In this case, the power spectrum is given by:

## $\frac{dP_b}{du} = \frac{d^2N_{\gamma}}{dtdu}E = P_b\left[2Z(1-u)\right]$

Where $P_b$ is the un-attenuated Brehmsstrahlung power, given below by:

## $P_b = \int_{0}^{1} \frac{dP_b}{du}du =\left[ \frac{2\alpha}{3\sqrt{3}\ell} \right] \left[ \frac{eV_0}{m_e c^2} \right] ZP_0$ 

where $P_0 = I_0 V_0$ is the electron beam power

### More Assumptions

The X-rays which reach the HV personnel are determined by the type and thickness of shielding between the electrodes and the personnel. We assume for this that:

#### Assumption 2: Max Power
We assume that the max power of the X-rays is generated when the HV is operated at the maximum voltage across the Niobium electrodes: 60 kV (aka, $V_0 = 60,000$ Volts)

#### Assumption 3: Max Current
We assume that we're operating at the maximum current output of the circuit: 1 mA

#### Assumption 4: Characteristic X-ray power
We assume that the Characteristic X-rays (**Characteristic X-rays**: X-rays emitted when outer-shell electrons fill a vacancy in the inner shell of an atom, releasing X-rays in a pattern that is "characteristic" to each element) are all emitted at the K-edge (**K-edge**: a sudden increase in X-ray absorption that occurrs when the energy of the X-rays is just above the binding energy of the innermost electron shell of the atoms interacting with the photons) energy with intensity that is a factor of 10 greater than the total un attenuated Brehmsstrahlung power ($P_c = 10P_b$ at $E_K$)

#### Assumption 5: X-ray penetration through HV vacuum chamber
We assume that to reach the personnel, X-rays must penetrate an equivalent thickness of 5mm Pyrex glass whose attenuation length varies with the energy of the incident X-ray


### Comments on these assumptions

### X-ray power spectrum at the surface of the Pyrex glass

At the outer surface of the glass, the power spectrum has the following form:

## $\frac{dP_a}{du}=\left[ \frac{dP_b}{du}+P_c\delta(u-u_K) \right] e^{-\frac{\Delta t}{t_0(E)}}$

Where the parameters have the following definitions:
* $P_a$: The attenuated X-ray power per unit relative energy
* $\delta(x)$: Dirac delta function
* $u_K$: $u_K = \frac{E_K}{eV_0}$ is the K-edge X-ray energy relative to the voltage between the HV electrodes
* $\Delta t$: The thickness of the vacuum chamber viewports (5mm)
* $t_0(E)$: Attenuation length for an X-ray of energy $E$

We can utilize the algebraic expression for $\frac{dP_b}{du}$, but will still need to implement an integral to obtain the attenuated spectrum. The integration will be performed over u, which can range from 0 to 1. 

### Estimating the Absorbed Dose

Once we know the attenuation, we can estimate the abosrbed dose ($M_{abs}$) with:

## $M_{abs} \leq \left< P \right> \left[ \left( \frac{I_{dis}}{I_{max}} \right)t_d \right] \left[ \frac{\Omega}{4\pi} \right] \left[ \frac{f_T}{m_T} \right] \left[ \omega_R \omega_T \right] $

Where the parameters have the following definitions:
* $\left<P\right> = \int_{0}^{1} (\frac{dP_a}{du})du$ is the average X-ray power
* $\frac{I_{dis}}{I_0}$: the discharge current relative to the maximum current
* $t_d$: the average duration of the discharge
* $\Omega$: The solid angle covered by the personnel
* $f_T$: fraction of X-rays absorbed by tissue
* $m_T$: mass of tissue
* $\omega_R$: Radiation weighting factor
* $\omega_T$: Tissue weighting factor

### Assmuption 6: Information from previous electrode conditioning

Experience with previous electrode conditioning implies that:
* $t_d \leq 1$; discharges only occur during electrode conditioning
* X-rays only exit the vacuum chamber through one of the viewports, limiting $\frac{\Omega}{4\pi} \approx \frac{1}{6}$ 
* X-rays are completely absorbed when incident on a person: $f_T = 1$
* The tissue weighting factor is assumed to be $\omega_T=0.2$, the mass of the tissue is assumed to be $m_T=50$ kg, and that radiation weigthing factor is assumed to be $\omega_R=1.0$.

## Calculations for estimating the dose

```python
#Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
```

```python
#We will attempt to reproduce the results from JTS's hazard mitigation document. Here, we define constants
V_0 = 30000 # Volts
I_0 = 0.001 # Amps
E_K = 19000 # eV
alpha = 1/137 # unitless
ell = 6 # assuming unitless
m_e = 0.511 # MeV/c^2
Z = 41 # Atomic number for discharge material (Nyobium in this case)
c = 1 # units of c
q_e = 1 # units of e
E_0 = q_e*V_0 # eV
delta_t = 5 # thickness of vacuum viewports, units of mm
u_K = E_K / (q_e*V_0) # relative energy of K-edge X-rays
```

```python
#Calculate the characteristic X-ray power, which is 10*P_b
prefactor = (2 * alpha) / (3 * np.sqrt(3) * ell ) # Unitless
scaled_E = (E_0) / (m_e * 1E6) # Unitless
scaled_P = Z * I_0 * V_0 # A*V = Watts
P_c = 10 * prefactor * scaled_E * scaled_P # Watts

# This correctly reproduces the characteristic X-rays for Niobium @ 30 kV, see HV mitigation doc pg. 10
print("P_c = ", P_c, "W")
print("The intensity of the characteristic X-rays is: ", 1000*P_c, "mW")
```

```python
# Read in the mass attenuation information and turn it into an attenuation length
colnames=["Photon Energy [MeV]","Total Attenuation W/ Coherent Scattering [cm^2/g]","Total Attenuation W/O Coherent Scattering [cm^2/g]"]
df = pd.read_csv("Mass Attenuation Data.txt", names=colnames, sep = " ") # read data file, delimiter is a space
pyrex_dens = 2.23 # density of Pyrex in g/cm^3
df["Coherent Attenuation Length [mm]"] = 10 / (df["Total Attenuation W/ Coherent Scattering [cm^2/g]"] * pyrex_dens)
df["Incoherent Attenuation Length  [mm]"] = 10 / (df["Total Attenuation W/O Coherent Scattering [cm^2/g]"] * pyrex_dens)
df["Energy [eV]"] = df["Photon Energy [MeV]"] * 1E6
df

# Plot the mass attenuation coefficient to check it with what XCOM gives us
plt.plot(df["Energy [eV]"]/1000,df["Total Attenuation W/ Coherent Scattering [cm^2/g]"])
plt.yscale("log")
plt.xscale("log")
plt.ylabel("Mass attenuation coef. (cm^2/g)")
plt.xlabel("X-Ray Energy (keV)")
plt.savefig("Mass Attenuation vs X-Ray Energy.png")
plt.show()

# Plot the attenuation length vs the x-ray energy and see if it matches Figure 4 from the HV Mitigation Doc
plt.plot(df["Energy [eV]"]/1000,df["Coherent Attenuation Length [mm]"])
plt.xlabel("X-Ray energy (keV)")
plt.ylabel("Attenuation length in PYREX 7740 (mm)")
plt.yscale("log")
plt.savefig("Attenuation Length vs X-Ray Energy.png")
```

```python
# Define the delta function approximation suggested by Nick 
def delta(x, alpha=0.0008):
    alpha_sq = alpha*alpha
    # Delta function approximation reference from Wolfram: https://mathworld.wolfram.com/DeltaFunction.html
    return (1 / np.pi) * alpha / (alpha_sq + x*x) 
```

```python
# Defining the X-ray power spectrum, Eq'n (4) from HV Mitigation Doc
def power_spectrum(u, t0_E_value): # Watts
    
    # First bracketed term in P_b, which is Eq'n (3) in HV Mitigation Doc
    prefactor = (2*alpha)/(3*np.sqrt(3)*ell) # Unitless
    
    # Second bracketed term in P_b, which is Eq'n (3) in HV Mitigation Doc
    scaled_E=(E_0)/(m_e*1E6) # Unitless
    
    # Z*P_0 term in P_b, which is Eq'n (3) in HV Mitigation Doc
    scaled_P = Z*I_0*V_0 # Watts
    
    # The bracketed term in the dPb_du expression, Equation (2) in the HV Mitigation Doc
    energy_factor = 2*Z*(1-u) # Unitless
    
    # Algebraical expression of Equation (2) in code
    dPb_du = prefactor*scaled_E*scaled_P*energy_factor # Watts
    
    # Delta function term from K-edge X-rays
    delta_P_c = P_c*delta(u-u_K) # Watts
    
    # Exponential attenuation term
    exp_factor = np.exp(-delta_t / t0_E_value) # unitless
    
    # Return Eq'n (4) on page 9 of HV Mitigation Doc
    return (dPb_du + delta_P_c) * exp_factor
```

```python
# Set up some initial index i and the list of dPa/du values
"""The manual indexing seems tedious, but I had trouble indexing over the contents
of the Energy [eV] column and grabbing the Coherent Attenuation Length [mm] that
is located in the same row. This works, but its a less clean code."""
i=0
dPa_values = []

# Loop over energies to pick out attenuation lengths
for i in range(len(df["Energy [eV]"])):
    E = df["Energy [eV]"][i]
    u=E/E_0 # relative energy, unitless
    t0_E = df["Coherent Attenuation Length [mm]"][i]

    # Calculate the result of Eq'n (4) and add it to a list
    dPa_regular = power_spectrum(u, t0_E)
    dPa_values.append(dPa_regular)
    i+=1 

# Add the result of Eq'n (4) to the dataframe and rename it
df = df.assign(dPa_du = dPa_values)
df["dPa_du"]
```

```python
# Set up figure size
plt.figure(figsize=(12,8))

# Plot energy in keV vs the output of Eq'n (4) from the HV Mitigation doc. Y-axis log scale, limits, and labels
plt.plot(df["Energy [eV]"]/1000, df["dPa_du"], marker="o", linestyle="--")
plt.yscale("log")
plt.xlabel("Energy (keV)")
plt.xlim(0,31)
plt.ylim(10**(-8),10**(-2))
plt.ylabel("Attenuated X-ray power (W)")

# Add horizontal line to mark the maximum data point on Figure 3 from HV Mitigation doc.
# Add K-edge energy (19 keV) and the location of the peak on Figure 3 from HV Mitigation doc.
plt.axvline(x=19, color="r", label="E=19 keV")
plt.axvline(x=24, color="m", linestyle="--", label="E=24 keV")
plt.axhline(y=0.0018, color="g", linestyle=":", label="1.8 mW")

# Try to enforce minor tick spacing
ax = plt.gca()
ax.yaxis.set_major_locator(ticker.LogLocator(base=10, numticks=10))
ax.minorticks_on()

# Print the legend and save output to png
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
plt.savefig("Attenuated X-Ray Spectrum.png")
```
